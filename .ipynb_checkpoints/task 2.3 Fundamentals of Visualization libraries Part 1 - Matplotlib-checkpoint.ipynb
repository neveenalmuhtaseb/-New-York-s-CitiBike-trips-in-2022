```python
# Sort the DataFrame by date
df_temp = df_temp.sort_index()

# average duplicate values  dates (optional: average them if needed)
df_temp = df_temp.groupby(df_temp.index).mean(numeric_only=True)

plt.figure(figsize=(10,6))
plt.title('Temperatures daily for 2022')
plt.plot(df_temp.index.values, df_temp['avgTemp'].values)
plt.xlabel('Date')
plt.ylabel('Average Temperature')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Groupby the count of the trips 

df_group = pd.DataFrame(df.groupby(['date'])['ride_id'].count()).reset_index()
```

```python
df_group.rename(columns = {'ride_id':'bike_rides_daily'}, inplace = True)
df = df.merge(df_group, on = "date", how = 'outer', indicator = True)
print(df['_merge'].value_counts(dropna = False))
print("Shape of January 1st is", df[df['date'] == '2022-01-01'].shape) # Check 
print("Shape of January 2nd is", df[df['date'] == '2022-01-02'].shape) # Second check 

df_temp = df.set_index('date')

print(df_temp.columns)
```

# Create a line chart of bike trip counts and temperatures plotted on a dual axis.

```python
# Create figure and axis objects with subplots() and details

fig,ax = plt.subplots(figsize=(10, 5))

# Make a plot of the bike rides
ax.plot(df_temp['bike_rides_daily'],color = "navy")
ax.set_xlabel("Year 2022", fontsize = 14) # set x-axis label
ax.set_ylabel("Bike rides daily",color = "navy", fontsize = 14) # set y-axis label

# Here is where we create the dual axis - initialize a twin object for two different y-axis on the sample plot
ax2 = ax.twinx()

# Make a plot with the temperatues on the y-axis using second axis object
ax2.plot(df_temp["avgTemp"], color = "red")
ax2.set_ylabel("Average temperatures", color = "red",fontsize=14)
plt.title('Temperature and trips in 2018', fontsize = 18)
plt.show()
```

```python
plt.style.use('ggplot')
```

```python
# Create figure and axis objects with subplots() and details

fig,ax = plt.subplots(2, figsize =(8, 6))

# Make a plot of the bike rides
ax[0].plot(df_temp['bike_rides_daily'],color = "navy")
ax[0].set_ylabel("Bike rides daily",color = "navy", fontsize = 14) # set y-axis label

# Make a plot with the temperatues on the y-axis using second axis object
ax[1].plot(df_temp["avgTemp"], color = "red")
ax[1].set_ylabel("Average temperatures", color = "red",fontsize=14)
plt.suptitle('Temperature and trips in 2022', fontsize = 18)

plt.show()
```

```python
# save the plot as a file
fig.savefig('trips_temperatures_dual_line.jpg',
            format='jpeg',
            dpi=100,
            bbox_inches='tight')
```

In this code, I used Matplotlibâ€™s **Object-Oriented (OO)** interface to create a figure with two vertically stacked subplots that display daily bike rides and average temperatures in 2022.

the **Object-Oriented paradigm** in Matplotlib, involves explicitly creating and manipulating Figure and Axes objects. This approach offers greater control and is ideal for creating complex, well-organized visualizations like multiple subplots.

```python
# First, convert to datetime if not already
df['started_at'] = pd.to_datetime(df['started_at'])
df['ended_at'] = pd.to_datetime(df['ended_at'])

# Create a new 'tripduration' column in minutes (or seconds if you prefer)
df['tripduration'] = (df['ended_at'] - df['started_at']).dt.total_seconds() / 60
```

```python
from scipy.stats import gaussian_kde

# Histogram + KDE
fig, ax = plt.subplots(figsize=(8, 5))

# Filter out extreme trip durations (e.g., >120 mins) if desired
filtered = df[df['tripduration'] <= 120]['tripduration']

counts, bins, patches = ax.hist(filtered, bins=30, color='skyblue', alpha=0.7, density=True, label='Histogram')

kde = gaussian_kde(filtered)
x_vals = np.linspace(filtered.min(), filtered.max(), 1000)
ax.plot(x_vals, kde(x_vals), color='darkblue', lw=2, label='KDE Curve')

ax.set_title('Trip Duration Histogram with KDE Curve')
ax.set_xlabel('Trip Duration (minutes)')
ax.set_ylabel('Density')
ax.legend()

plt.show()
```
